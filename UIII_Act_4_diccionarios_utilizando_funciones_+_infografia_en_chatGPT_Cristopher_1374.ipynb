Crea una infografía moderna y futurista sobre un programa en Python de diccionario de panaderos. Incluye funciones agregar_panadero(), listar_panaderos(), buscar_panadero(), borrar_panadero() y actualizar_panadero(), usando while True, if elif y for key, value. Diseño tecnológico azul oscuro con estilo IA, iconos de panadería, bloques de código Python, menú interactivo y ejemplos de diccionarios con 4 panaderos registrados.

```python
# Inicializamos el diccionario vacío
diccionario_panadero = {}

# --- FUNCIONES ---

def agregar_panadero():
    nombre = input("Nombre del panadero: ").capitalize()
    especialidad = input(f"Especialidad de {nombre} (ej. Masas madre, Facturas): ")
    diccionario_panadero[nombre] = especialidad
    print(f"¡{nombre} ha sido agregado con éxito!")

def listar_panaderos():
    if not diccionario_panadero:
        print("\nEl diccionario está vacío.")
    else:
        print("\n--- Lista de Panaderos ---")
        # Uso de for key, value (clave, valor)
        for nombre, especialidad in diccionario_panadero.items():
            print(f"Panadero: {nombre} | Especialidad: {especialidad}")

def buscar_panadero():
    nombre = input("Nombre del panadero a buscar: ").capitalize()
    if nombre in diccionario_panadero:
        print(f"Encontrado -> {nombre}: {diccionario_panadero[nombre]}")
    else:
        print("Ese panadero no se encuentra en el registro.")

def borrar_panadero():
    nombre = input("Nombre del panadero a eliminar: ").capitalize()
    if nombre in diccionario_panadero:
        del diccionario_panadero[nombre]
        print(f"El panadero {nombre} ha sido eliminado.")
    else:
        print("No se pudo eliminar: el nombre no existe.")

def actualizar_panadero():
    nombre = input("Nombre del panadero que deseas actualizar: ").capitalize()
    if nombre in diccionario_panadero:
        nueva_especialidad = input(f"Nueva especialidad para {nombre}: ")
        diccionario_panadero[nombre] = nueva_especialidad
        print("Datos actualizados correctamente.")
    else:
        print("El panadero no existe.")

# --- MENÚ INTERACTIVO ---

# Datos iniciales (4 registros requeridos)
diccionario_panadero["Juan"] = "Pan Francés"
diccionario_panadero["Maria"] = "Pastelería"
diccionario_panadero["Pedro"] = "Masas Madre"
diccionario_panadero["Ana"] = "Bollería"

while True:
    print("\n--- MENÚ DE GESTIÓN PANADERO ---")
    print("1. Agregar panadero")
    print("2. Listar panaderos")
    print("3. Buscar panadero")
    print("4. Borrar panadero")
    print("5. Actualizar panadero")
    print("6. Salir")

    opcion = input("Seleccione una opción (1-6): ")

    if opcion == "1":
        agregar_panadero()
    elif opcion == "2":
        listar_panaderos()
    elif opcion == "3":
        buscar_panadero()
    elif opcion == "4":
        borrar_panadero()
    elif opcion == "5":
        actualizar_panadero()
    elif opcion == "6":
        print("Saliendo del programa... ¡Buen día!")
        break
    else:
        print("Opción no válida, intente de nuevo.")
```
