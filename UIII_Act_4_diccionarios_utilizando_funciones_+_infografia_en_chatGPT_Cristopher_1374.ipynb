Crea una infografía moderna y futurista sobre un programa en Python de diccionario de panaderos. Incluye funciones agregar_panadero(), listar_panaderos(), buscar_panadero(), borrar_panadero() y actualizar_panadero(), usando while True, if elif y for key, value. Diseño tecnológico azul oscuro con estilo IA, iconos de panadería, bloques de código Python, menú interactivo y ejemplos de diccionarios con 4 panaderos registrados.

```python
# Inicializamos el diccionario vacío
diccionario_panadero = {}

# Datos iniciales (4 registros para cumplir con el requisito)
diccionario_panadero["Juan"] = "Pan Francés"
diccionario_panadero["Maria"] = "Pastelería"
diccionario_panadero["Pedro"] = "Masas Madre"
diccionario_panadero["Ana"] = "Bollería"

# --- FUNCIONES ---

def agregar_panadero():
    nombre = input("\n> Ingrese nombre del panadero: ").capitalize()
    especialidad = input(f"> Especialidad de {nombre}: ")
    diccionario_panadero[nombre] = especialidad
    print(f"✅ Panadero {nombre} agregado.")

def listar_panaderos():
    if not diccionario_panadero:
        print("\n[!] El registro está vacío.")
    else:
        print("\n" + "="*40)
        print(f"{'NOMBRE':<20} | {'ESPECIALIDAD':<15}")
        print("-" * 40)
        for nombre, especialidad in diccionario_panadero.items():
            print(f"{nombre:<20} | {especialidad:<15}")
        print("="*40)

def buscar_panadero():
    nombre = input("\n> Nombre a buscar: ").capitalize()
    if nombre in diccionario_panadero:
        print(f"🔍 Resultado: {nombre} se especializa en {diccionario_panadero[nombre]}.")
    else:
        print("❌ No se encontró ese panadero.")

def borrar_panadero():
    nombre = input("\n> Nombre a eliminar: ").capitalize()
    if nombre in diccionario_panadero:
        del diccionario_panadero[nombre]
        print(f"🗑️ {nombre} ha sido eliminado del sistema.")
    else:
        print("❌ El nombre no existe.")

def actualizar_panadero():
    nombre = input("\n> Nombre a actualizar: ").capitalize()
    if nombre in diccionario_panadero:
        nueva_esp = input(f"> Nueva especialidad para {nombre}: ")
        diccionario_panadero[nombre] = nueva_esp
        print("📝 Datos actualizados.")
    else:
        print("❌ Panadero no encontrado.")

# --- BUCLE PRINCIPAL CON MENÚ TIPO TABLA ---

while True:
    print("\n+-----------------------------------------+")
    print("|      SISTEMA DE GESTIÓN: PANADERÍA      |")
    print("+--------+--------------------------------+")
    print("| OPCIÓN |           DESCRIPCIÓN          |")
    print("+--------+--------------------------------+")
    print("|   1    | Agregar Panadero               |")
    print("|   2    | Listar Panaderos               |")
    print("|   3    | Buscar Panadero                |")
    print("|   4    | Borrar Panadero                |")
    print("|   5    | Actualizar Panadero            |")
    print("|   6    | Salir del Sistema              |")
    print("+--------+--------------------------------+")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        agregar_panadero()
    elif opcion == "2":
        listar_panaderos()
    elif opcion == "3":
        buscar_panadero()
    elif opcion == "4":
        borrar_panadero()
    elif opcion == "5":
        actualizar_panadero()
    elif opcion == "6":
        print("\nCerrando el horno... ¡Hasta luego!")
        break
    else:
        print("\n[!] Opción inválida. Intente de nuevo.")
```
