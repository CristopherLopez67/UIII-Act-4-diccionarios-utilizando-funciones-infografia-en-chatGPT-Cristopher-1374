Crea una infografía moderna y futurista sobre un programa en Python de diccionario de panaderos. Incluye funciones agregar_panadero(), listar_panaderos(), buscar_panadero(), borrar_panadero() y actualizar_panadero(), usando while True, if elif y for key, value. Diseño tecnológico azul oscuro con estilo IA, iconos de panadería, bloques de código Python, menú interactivo y ejemplos de diccionarios con 4 panaderos registrados.

```python
# Inicializamos el diccionario vacío
Diccionario_panadero = {}

# --- FUNCIÓN PARA AGREGAR DATOS INICIALES (4 Registros) ---
def carga_inicial():
    Diccionario_panadero["Julian"] = {
        "Pan": "Masa Madre", "Horno": "Leña", "Produccion": 50,
        "Local": "La Tahona", "Receta": "Fermentación 48h", "Horario": "02:00 AM"
    }
    Diccionario_panadero["Marta"] = {
        "Pan": "Baguette", "Horno": "Eléctrico", "Produccion": 120,
        "Local": "Pan de Oro", "Receta": "Harina Francesa", "Horario": "03:30 AM"
    }
    Diccionario_panadero["Roberto"] = {
        "Pan": "Ciabatta", "Horno": "Convección", "Produccion": 80,
        "Local": "El Trigal", "Receta": "Aceite de Oliva", "Horario": "01:00 AM"
    }
    Diccionario_panadero["Elena"] = {
        "Pan": "Facturas", "Horno": "Rotativo", "Produccion": 300,
        "Local": "Dulce Trigo", "Receta": "Almíbar Casero", "Horario": "04:00 AM"
    }

# --- FUNCIONES DEL SISTEMA ---

def agregar_panadero():
    nombre = input("\nNombre del Panadero: ").capitalize()
    datos = {
        "Pan": input("Tipo de pan: "),
        "Horno": input("Horno utilizado: "),
        "Produccion": input("Producción diaria (uds): "),
        "Local": input("Nombre de la panadería: "),
        "Receta": input("Receta secreta: "),
        "Horario": input("Horario de madrugada: ")
    }
    Diccionario_panadero[nombre] = datos
    print(f"✅ ¡{nombre} ha sido registrado en el gremio!")

def listar_panaderos():
    if not Diccionario_panadero:
        print("\nNo hay panaderos registrados.")
    else:
        print("\n" + "="*115)
        print(f"{'PANADERO':<12} | {'PAN':<15} | {'HORNO':<12} | {'PROD.':<6} | {'PANADERÍA':<15} | {'RECETA':<18} | {'HORARIO'}")
        print("-" * 115)
        # Uso de for clave, valor
        for nombre, info in Diccionario_panadero.items():
            print(f"{nombre:<12} | {info['Pan']:<15} | {info['Horno']:<12} | {info['Produccion']:<6} | {info['Local']:<15} | {info['Receta']:<18} | {info['Horario']}")
        print("="*115)

def buscar_panadero():
    nombre = input("\nNombre del panadero a buscar: ").capitalize()
    if nombre in Diccionario_panadero:
        p = Diccionario_panadero[nombre]
        print(f"\n🔍 DATOS DE {nombre.upper()}:")
        for k, v in p.items():
            print(f"- {k}: {v}")
    else:
        print("❌ El panadero no existe.")

def borrar_panadero():
    nombre = input("\nNombre del panadero a eliminar: ").capitalize()
    if nombre in Diccionario_panadero:
        del Diccionario_panadero[nombre]
        print(f"🗑️ Registro de {nombre} eliminado.")
    else:
        print("❌ No se encontró el registro.")

def actualizar_panadero():
    nombre = input("\nNombre del panadero a actualizar: ").capitalize()
    if nombre in Diccionario_panadero:
        print("Deje en blanco para mantener el dato actual.")
        for k in Diccionario_panadero[nombre]:
            nuevo_valor = input(f"Nuevo valor para {k} (Actual: {Diccionario_panadero[nombre][k]}): ")
            if nuevo_valor:
                Diccionario_panadero[nombre][k] = nuevo_valor
        print("📝 Registro actualizado con éxito.")
    else:
        print("❌ El panadero no existe.")

# --- EJECUCIÓN ---

carga_inicial()

while True:
    print("\n+---------------------------------------+")
    print("|      GESTIÓN DE PANADEROS EXPERTOS    |")
    print("+--------+------------------------------+")
    print("| OPCIÓN |         DESCRIPCIÓN          |")
    print("+--------+------------------------------+")
    print("|   1    | Agregar nuevo Panadero       |")
    print("|   2    | Listar todos los Panaderos   |")
    print("|   3    | Buscar por nombre            |")
    print("|   4    | Borrar registro              |")
    print("|   5    | Actualizar información       |")
    print("|   6    | Salir del programa           |")
    print("+--------+------------------------------+")

    opcion = input("Seleccione (1-6): ")

    if opcion == "1":
        agregar_panadero()
    elif opcion == "2":
        listar_panaderos()
    elif opcion == "3":
        buscar_panadero()
    elif opcion == "4":
        borrar_panadero()
    elif opcion == "5":
        actualizar_panadero()
    elif opcion == "6":
        print("\n¡Apagando los hornos! Hasta pronto.")
        break
    else:
        print("\n⚠ Opción inválida.")
```
